```python
def centroid_mean(X,idx,K):
    m, n = X.shape[0],X.shape[1]
    centroids = np.zeros((K,n))
    count = np.zeros((K,1))
    
    for i in range(m):
        index = int((idx[i]-1)[0])
        centroids[index,:]+=X[i,:]
        count[index]+=1
    
    return centroids/count
    
```

```python
def centro_mean(X,idx,K):
    list1=[]
    count=[]
    da=[]
    for j in range(1,K+1):
        ct= [(int(idx[i])==j) for i in range(X.shape[0])]
        kol=np.where(ct)[0]
        list1.append(kol)
        count.append(len(kol))
    for l in range(K):
        po=list1[l]
        value=0
        for m in po:
            value+=X[m]
        da.append(value)
    return np.array(np.matrix(da))/np.array(count).reshape(K,1)
 
```

Above function is more Basic function for calculating means

```python
def closest_centroid1(X,centroid):
    K=centroid.shape[0]
    idx=np.zeros((X.shape[0],1))
    tem=np.zeros((centroid.shape[0],1))
    for i in range(X.shape[0]):
        for j in range(K):
            dist=X[i,:]-centroid[j,:]
            length=np.sum(dist**2)
            tem[j]=length
            #print("value of x",X[i,:],"!","Distances",length)
        idx[i]=np.argmin(tem)+1
        #print("index we choose",idx[i])
    return idx
```

```python
def plotKmeans(X, centroids, idx, K, num_iters):
    m,n = X.shape[0],X.shape[1]
    fig, ax = plt.subplots(nrows=num_iters,ncols=1,figsize=(6,36))
    for i in range(num_iters):    
        # Visualisation of data
        color = "rgb"
        for k in range(1,K+1):
            grp = (idx==k).reshape(m,1)
            ax[i].scatter(X[grp[:,0],0],X[grp[:,0],1],c=color[k-1],s=15)

        # visualize the new centroids
        ax[i].scatter(centroids[:,0],centroids[:,1],s=120,marker="x",c="black",linewidth=3)
        title = "Iteration Number " + str(i)
        ax[i].set_title(title)
        
        # Compute the centroids mean
        centroids = centro_mean(X, idx, K)
        
        # assign each training example to the nearest centroid
        idx = closest_centroid1(X, centroids)
    
    plt.tight_layout()
```

# Random Initilization

```python
def randominitilization(X, K):
    m,n = X.shape[0], X.shape[1]
    centroids = np.zeros((K,n))
    for i in range(K):
        centroids[i] = X[np.random.randint(0,m+1),:]
        
    return centroids
```

```python
initial_centroids_new=randominitilization(X,3)
```

```python
idx=closest_centroid1(X,initial_centroids_new)
```

```python
m,n = X.shape[0],X.shape[1]
plotKmeans(X, initial_centroids_new,idx, K,10)
```

```python
X2 = (X/255).reshape(128*128,3)
```

```python
X2.shape
```

```python
def runKmeans(X, initial_centroids,num_iters,K):
    
    idx = closest_centroid1(X, initial_centroids)
    
    for i in range(num_iters):
        
        # Compute the centroids mean
        centroids = centro_mean(X, idx, K)

        # assign each training example to the nearest centroid
        idx = closest_centroid1(X, initial_centroids)

        
    return centroids, idx
```

```python
K2 = 16
num_iters = 10
initial_centroids2 = randominitilization(X2, K2)
centroids2, idx2 = runKmeans(X2, initial_centroids2, num_iters,K2)
```

```python
m2,n2 = X.shape[0],X.shape[1]
X2_recovered = X2.copy()
for i in range(1,K2+1):
    X2_recovered[(idx2==i).ravel(),:] = centroids2[i-1]

# Reshape the recovered image into proper dimensions
X2_recovered = X2_recovered.reshape(128,128,3)

# Display the image
import matplotlib.image as mpimg
fig, ax = plt.subplots(1,2)
ax[0].imshow(X2.reshape(128,128,3))
ax[1].imshow(X2_recovered)
```
